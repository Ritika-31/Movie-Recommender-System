```python
import numpy as np
import pandas as pd
```

```python
movies=pd.read_csv('tmdb_5000_movies.csv')
credits=pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies=movies.merge(credits,on='title')
```

```python
movies.head(1)
```

```python
movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
        return L
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast']=movies['cast'].apply(convert3)
```

```python
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
```

```python
 movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
def collapse(L):
    L1=[]
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast']=movies['cast'].apply(collapse)
# movies['keywords']=movies['keywords'].apply(collapse)
# movies['crew']=movies['crew'].apply(collapse)
# movies['genres']=movies['genres'].apply(collapse)
```

```python
movies['crew']=movies['crew'].apply(collapse)
```

```python
movies['tags']=movies['overview']+movies['cast']+movies['crew']+movies['genres']
```

```python
movies.head(1)
```

```python
new_df=movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
# def join_tags(tags_list):
    # return " ".join(tags_list)

# new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
# def join_tags(tags):
#     if isinstance(tags, list):
#         return " ".join(tags)
#     else:
#         return tags

# new_df['tags'] = new_df['tags'].apply(join_tags)
def join_tags(tags):
    if isinstance(tags, list):
        return " ".join(tags)
    else:
        return tags

new_df.loc[:, 'tags'] = new_df['tags'].apply(join_tags)
```

```python
new_df['tags'][0]
```

```python
# new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
# def lowercase_tags(tag):
#     if isinstance(tag, str):
#         return tag.lower()
#     else:
#         return tag

# new_df['tags'] = new_df['tags'].apply(lowercase_tags)
def lowercase_tags(tag):
    if isinstance(tag, str):
        return tag.lower()
    else:
        return tag

new_df.loc[:, 'tags'] = new_df['tags'].apply(lowercase_tags)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
# cv.fit_transform(new_df['tags']).toarray()
# Drop rows where 'tags' is NaN
new_df = new_df.dropna(subset=['tags'])

# Apply CountVectorizer
from sklearn.feature_extraction.text import CountVectorizer 

cv = CountVectorizer()
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
['loved','loving','love']
['love','love','love']
```

```python
ps.stem('dance')
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity=cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
sorted(similarity[0],reverse=True)
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movie_list:
        print(new_df.iloc[i[0]].title)
```
